```python
import numpy as np
import matplotlib.pyplot as plt

# M√©todo de Euler vectorizado
def euler_method_vectorized(f, t0, y0, h, t_end):
    t_values = np.arange(t0, t_end + h, h)
    y_values = np.zeros_like(t_values)
    y_values[0] = y0

    for i in range(1, len(t_values)):
        y_values[i] = y_values[i - 1] + h * f(t_values[i - 1], y_values[i - 1])

    return t_values, y_values

# Problema a
def f_a(t, y):
    return t * np.exp(3 * t) - 2 * y

t_values_a, y_values_a = euler_method_vectorized(f_a, 0, 0, 0.5, 1)

# Problema b
def f_b(t, y):
    return 1 + (t - y)**2

t_values_b, y_values_b = euler_method_vectorized(f_b, 2, 1, 0.5, 3)

# Problema c
def f_c(t, y):
    return 1 + y / t

t_values_c, y_values_c = euler_method_vectorized(f_c, 1, 2, 0.25, 2)

# Problema d
def f_d(t, y):
    return np.cos(2 * t) + np.sin(3 * t)

t_values_d, y_values_d = euler_method_vectorized(f_d, 0, 1, 0.25, 1)

# Imprimir los resultados
print("Resultados del Problema a:")
for t, y in zip(t_values_a, y_values_a):
    print(f"t = {t:.2f}, y(t) = {y:.4f}")

print("\nResultados del Problema b:")
for t, y in zip(t_values_b, y_values_b):
    print(f"t = {t:.2f}, y(t) = {y:.4f}")

print("\nResultados del Problema c:")
for t, y in zip(t_values_c, y_values_c):
    print(f"t = {t:.2f}, y(t) = {y:.4f}")

print("\nResultados del Problema d:")
for t, y in zip(t_values_d, y_values_d):
    print(f"t = {t:.2f}, y(t) = {y:.4f}")

# Graficar los resultados
plt.figure(figsize=(12, 8))

# Gr√°fico para el problema a
plt.subplot(2, 2, 1)
plt.plot(t_values_a, y_values_a, marker='o', label='Problema a')
plt.title('Problema a: $y\' = te^{3t} - 2y$')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid(True)
plt.legend()

# Gr√°fico para el problema b
plt.subplot(2, 2, 2)
plt.plot(t_values_b, y_values_b, marker='o', label='Problema b')
plt.title('Problema b: $y\' = 1 + (t - y)^2$')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid(True)
plt.legend()

# Gr√°fico para el problema c
plt.subplot(2, 2, 3)
plt.plot(t_values_c, y_values_c, marker='o', label='Problema c')
plt.title('Problema c: $y\' = 1 + \\frac{y}{t}$')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid(True)
plt.legend()

# Gr√°fico para el problema d
plt.subplot(2, 2, 4)
plt.plot(t_values_d, y_values_d, marker='o', label='Problema d')
plt.title('Problema d: $y\' = \cos(2t) + \sin(3t)$')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# M√©todo de Euler vectorizado
def euler_method(f, t0, y0, h, t_end):
    t_values = np.arange(t0, t_end + h, h)
    y_values = np.zeros_like(t_values)
    y_values[0] = y0

    for i in range(1, len(t_values)):
        y_values[i] = y_values[i - 1] + h * f(t_values[i - 1], y_values[i - 1])

    return t_values, y_values

# Problema a: y' = y/t - (y/t)^2, 1 <= t <= 2, y(1) = 1, h = 0.1
def f_a(t, y):
    return y / t - (y / t) ** 2

t_values_a, y_values_a = euler_method(f_a, 1, 1, 0.1, 2)

# Problema b: y' = 1 + y/t + (y/t)^2, 1 <= t <= 3, y(1) = 0, h = 0.2
def f_b(t, y):
    return 1 + y / t + (y / t) ** 2

t_values_b, y_values_b = euler_method(f_b, 1, 0, 0.2, 3)

# Problema c: y' = -(y + 1)(y + 3), 0 <= t <= 2, y(0) = -2, h = 0.2
def f_c(t, y):
    return -(y + 1) * (y + 3)

t_values_c, y_values_c = euler_method(f_c, 0, -2, 0.2, 2)

# Problema d: y' = -5y + 5t^2 + 2t, 0 <= t <= 1, y(0) = 1/3, h = 0.1
def f_d(t, y):
    return -5 * y + 5 * t**2 + 2 * t

t_values_d, y_values_d = euler_method(f_d, 0, 1/3, 0.1, 1)

# Graficar los resultados
plt.figure(figsize=(12, 8))

# Gr√°fico para el problema a
plt.subplot(2, 2, 1)
plt.plot(t_values_a, y_values_a, 'o-', label='Aproximaci√≥n Euler')
plt.title('Problema a')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid(True)
plt.legend()

# Gr√°fico para el problema b
plt.subplot(2, 2, 2)
plt.plot(t_values_b, y_values_b, 'o-', label='Aproximaci√≥n Euler')
plt.title('Problema b')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid(True)
plt.legend()

# Gr√°fico para el problema c
plt.subplot(2, 2, 3)
plt.plot(t_values_c, y_values_c, 'o-', label='Aproximaci√≥n Euler')
plt.title('Problema c')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid(True)
plt.legend()

# Gr√°fico para el problema d
plt.subplot(2, 2, 4)
plt.plot(t_values_d, y_values_d, 'o-', label='Aproximaci√≥n Euler')
plt.title('Problema d')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

4. Aqu√≠ se dan las soluciones reales para los problemas de valor inicial en el ejercicio 3. Calcule el error real en las aproximaciones del ejercicio 3.

```python
import numpy as np

# Definir la funci√≥n y' y la soluci√≥n real
def f(t, y):
    return 1/t + (y/t)**2

def y_real(t):
    return t * np.tan(np.log(t))

# Valores iniciales
t0, y0 = 1, 0
h = 0.2
t_final = 3

# M√©todo de Euler
t_values = np.arange(t0, t_final + h, h)
y_values = [y0]

for t in t_values[:-1]:
    y = y_values[-1]
    y_next = y + h * f(t, y)
    y_values.append(y_next)

# Calcular el error real
errors = [abs(y_real(t) - y) for t, y in zip(t_values, y_values)]

# Imprimir resultados
for t, y, error in zip(t_values, y_values, errors):
    print(f"t = {t:.1f}, y_approx = {y:.6f}, y_real = {y_real(t):.6f}, error = {error:.6f}")
```

6.	Use el m√©todo de Taylor de orden 2 para aproximar las soluciones para cada uno de los siguientes problemas de valor inicial.

```python
import numpy as np
import matplotlib.pyplot as plt

# M√©todo de Taylor de orden 2
def taylor_order2(f, f_t, f_y, t0, y0, h, t_end):
    t_values = np.arange(t0, t_end + h, h)
    y_values = np.zeros_like(t_values)
    y_values[0] = y0

    for i in range(1, len(t_values)):
        t_n, y_n = t_values[i-1], y_values[i-1]
        f_n = f(t_n, y_n)
        y_values[i] = y_n + h * f_n + (h**2 / 2) * (f_t(t_n, y_n) + f_y(t_n, y_n) * f_n)

    return t_values, y_values

# Problema a: y' = t * e^(3t) - 2y, 0 <= t <= 1, y(0) = 0, h = 0.5
def f_a(t, y):
    return t * np.exp(3 * t) - 2 * y

def f_t_a(t, y):
    return np.exp(3 * t) * (3 * t + 1)

def f_y_a(t, y):
    return -2

t_values_a, y_values_a = taylor_order2(f_a, f_t_a, f_y_a, 0, 0, 0.5, 1)

# Problema b: y' = 1 + (t - y)^2, 2 <= t <= 3, y(2) = 1, h = 0.5
def f_b(t, y):
    return 1 + (t - y) ** 2

def f_t_b(t, y):
    return 2 * (t - y)

def f_y_b(t, y):
    return -2 * (t - y)

t_values_b, y_values_b = taylor_order2(f_b, f_t_b, f_y_b, 2, 1, 0.5, 3)

# Problema c: y' = 1 + y/t, 1 <= t <= 2, y(1) = 2, h = 0.25
def f_c(t, y):
    return 1 + y / t

def f_t_c(t, y):
    return -y / t**2

def f_y_c(t, y):
    return 1 / t

t_values_c, y_values_c = taylor_order2(f_c, f_t_c, f_y_c, 1, 2, 0.25, 2)

# Problema d: y' = cos(2t) + sin(3t), 0 <= t <= 1, y(0) = 1, h = 0.25
def f_d(t, y):
    return np.cos(2 * t) + np.sin(3 * t)

def f_t_d(t, y):
    return -2 * np.sin(2 * t) + 3 * np.cos(3 * t)

def f_y_d(t, y):
    return 0  # La funci√≥n f(t, y) no depende de y

t_values_d, y_values_d = taylor_order2(f_d, f_t_d, f_y_d, 0, 1, 0.25, 1)

# Graficar resultados
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Problema a
axs[0, 0].plot(t_values_a, y_values_a, 'o-', label='Taylor Order 2')
axs[0, 0].set_title("Problema a: y' = te^(3t) - 2y")
axs[0, 0].set_xlabel("t")
axs[0, 0].set_ylabel("y(t)")
axs[0, 0].legend()

# Problema b
axs[0, 1].plot(t_values_b, y_values_b, 'o-', label='Taylor Order 2')
axs[0, 1].set_title("Problema b: y' = 1 + (t - y)^2")
axs[0, 1].set_xlabel("t")
axs[0, 1].set_ylabel("y(t)")
axs[0, 1].legend()

# Problema c
axs[1, 0].plot(t_values_c, y_values_c, 'o-', label='Taylor Order 2')
axs[1, 0].set_title("Problema c: y' = 1 + y/t")
axs[1, 0].set_xlabel("t")
axs[1, 0].set_ylabel("y(t)")
axs[1, 0].legend()

# Problema d
axs[1, 1].plot(t_values_d, y_values_d, 'o-', label='Taylor Order 2')
axs[1, 1].set_title("Problema d: y' = cos(2t) + sen(3t)")
axs[1, 1].set_xlabel("t")
axs[1, 1].set_ylabel("y(t)")
axs[1, 1].legend()

plt.tight_layout()
plt.show()
```
